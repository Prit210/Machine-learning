### onverting date and time columns to a single datetime object


```python


df['ENTRY_DATE'] = pd.to_datetime(df['ENTRY_DATE'], errors='coerce')
df['ENTRY_DATE'] = pd.to_datetime(df['ENTRY_DATE'], errors='coerce')
df['EVENT_LCL_DATE'] = pd.to_datetime(df['EVENT_LCL_DATE'], format='%H:%M', errors='coerce').dt.time
```

### Extract the following attributes

```python
required_cols = ['ACFT_MAKE_NAME','LOC_STATE_NAME','ACFT_MODEL_NAME','RMK_TEXT','FLT_PHASE','EVENT_TYPE_DESC','FATAL_FLAG']
df_req = df[required_cols].copy()
```

#### Replacing all Fatal Flag missing values with the required output

```python
df_req['FATAL_FLAG'] = df_req['FATAL_FLAG'].fillna('No')
```

#### For the FLT_PHASE and ACFT_DMG_DESC , used mode to fill in the missing data

```python
for col in ['FLT_PHASE', 'ACFT_DMG_DESC']:
    mode_val = df[col].mode().iloc[0]
    df[col] = df[col].fillna(mode_val)
```

#### Verify if the missing values are replaced


```python

df_req.isnull().sum()
```

#### Check the number of observations


### removing unnessary data

```python
# --- Drop observations with missing aircraft names ---
df_req = df_req.dropna(subset=['ACFT_MAKE_NAME'])
```

```python
# --- Drop columns having >75% missing values in original df ---
threshold = len(df) * 0.75
df_req = df_req.loc[:, df_req.notnull().sum() >= threshold]
```
