```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

n = 100   # number of trials (reviews)
observed_success = 97

p_values = [0.90, 0.95, 0.97, 0.99]

num_samples = 100000

plt.figure(figsize=(12, 8))

for i, p in enumerate(p_values, 1):
    samples = np.random.binomial(n, p, size=num_samples)
    
    plt.subplot(2, 2, i)
    plt.hist(samples, bins=30, density=True, color="skyblue", edgecolor="black")
    plt.axvline(observed_success, color="red", linestyle="--", label="Observed 97/100")
    plt.title(f"Binomial(n={n}, p={p})")
    plt.xlabel("Number of Positives")
    plt.ylabel("Probability Density")
    plt.legend()

plt.tight_layout()
plt.show()
```
