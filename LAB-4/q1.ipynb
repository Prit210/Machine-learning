```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Parameters
num_rolls_list = range(1, 51)  # from 1 to 50 dice rolls
num_experiments_list = [100, 500, 1000, 5000, 10000, 50000, 100000]
output_dir = "die_histograms"
n_rolls=50
# Create output folder
os.makedirs(output_dir, exist_ok=True)

for n_experiments in num_experiments_list:
    rolls = np.random.randint(1, 7, size=(n_experiments, n_rolls))
    sums = np.sum(rolls, axis=1)
    # Compute mean and variance
    mean_val = np.mean(sums)
    var_val = np.var(sums)
    print(f"Rolls={n_rolls}=, Experiments={n_experiments} -> Mean={mean_val:.4f}, Variance={var_val:.4f}")
    # Plot histogram
    plt.figure(figsize=(8, 5))
    plt.hist(sums, bins=30, density=True, color="skyblue", edgecolor="black")
    plt.title(f"Histogram of {n_rolls} Rolls ({n_experiments} Experiments)\nMean={mean_val:.2f}, Var={var_val:.2f}")
    plt.xlabel("Sum of Rolls")
    plt.ylabel("Frequency")
    # Save histogram
    filename = os.path.join(output_dir, f"hist_{n_rolls}rolls_{n_experiments}exp.png")
    plt.savefig(filename)
    plt.close()
```
